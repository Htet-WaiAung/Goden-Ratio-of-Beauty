```python
import GoldenFace
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

def goldenFaceRatio(image):
  color_a = (255,255,0)
  color_b = (0,0,255)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawFaceCover(color_a)
  img1 = analysis.img
  cv2.waitKey(10)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawTZM(color_b)
  img2 = analysis.img
  cv2.waitKey(10)


  analysis = GoldenFace.goldenFace(image)
  analysis.drawTGSM(color_b)
  img3 = analysis.img
  cv2.waitKey(10)

  firstrow = np.concatenate((img1, img2, img3), axis=1)
  cv2_imshow(firstrow)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawVFM(color_b)
  img4 = analysis.img
  cv2.waitKey(10)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawTSM(color_b)
  img5 = analysis.img
  cv2.waitKey(10)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawLC(color_b)
  img6 = analysis.img
  cv2.waitKey(10)

  secondrow = np.concatenate((img4, img5, img6), axis=1)
  cv2_imshow(secondrow)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawLandmarks(color_b)
  img7 = analysis.img
  cv2.waitKey(10)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawLandmark(color_a)
  img8 = analysis.img
  cv2.waitKey(10)

  analysis = GoldenFace.goldenFace(image)
  analysis.drawMask(color_a)
  img9 = analysis.img
  cv2.waitKey(10)

  thirdrow = np.concatenate((img7, img8, img9), axis=1)
  cv2_imshow(thirdrow)

  analysis = GoldenFace.goldenFace(image)
  goldenRatio = analysis.geometricRatio()

  #text = "Golden Ratio: " + str(int(goldenRatio)) + " %."

  #image = cv2.putText(analysis.img, text, (0,400), cv2.FONT_HERSHEY_SIMPLEX, 1, color_a, 2)
  #cv2_imshow(analysis.img)
  #cv2.waitKey(10)

  return goldenRatio
```

```python
img = "/content/lisa.jpg"
```

```python
golden = goldenFaceRatio(img)
result = " Your Face Golden Ratio is : " + str(int(golden)) + " %."
result
```
